```python
#Creating the dataset
df = pd.read_csv("/content/ONLINE EDUCATION SYSTEM REVIEW (1).csv")
#Creating the scatter plot
plt.scatter(df['Age(Years)'],df['Performance in online'])
#Adding the aesthetics
plt.title('Gender vs Age in ONline EDucation SYstem ')
plt.xlabel('Age')
plt.ylabel('Performance')
#Show the plot
plt.show()
```

```python
import seaborn as sns
sns.set_style('whitegrid')
sns.barplot(x='Performance in online',y='Age(Years)',data=df).set_title('Performance in Online Vs Age')
```

```python
plt.xlabel('Level of Education')

plt.ylabel('Count')

plt.title('Histogram of Level of Education')
plt.hist(df['Level of Education'])


```

```python
corr = df[['Gender', 'Age(Years)', 'Time spent on social media (Hours)', 'Home Location']].corr()

sns.heatmap(corr, annot=True)

plt.title('Gender vs Age on Time spent on Social Media')

plt.show()
```

```python
data = df[['Number of Subjects', 'Age(Years)', 'Study time (Hours)']]

sns.pairplot(data)

plt.show()
```
